```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path2csv = './csv/'#'/Users/klara/Downloads/'
path2plots = './plots/convergence/'
```

# preparation of data

```python
def csv_to_df(path2csv):
    dff_all_from_csv = []
    print(path2csv)
    
    for root, dirs, files in os.walk(path2csv):
        print(root, files)
        for file in files:
            if not file.endswith('.csv'):
                continue
            df = pd.read_csv(os.path.join(root, file), skiprows=range(0,6))
            df.set_index(['[run number]', '[step]'], inplace=True)
            dff_all_from_csv.append(df)
    return dff_all_from_csv
```

```python
dff_all_from_csv = csv_to_df(path2csv)
```

# display results wrt. # steps until convergence

```python
def plot_run_length(df):
    # determine type of grammar spread
    infection_type = df['update-algorithm'].iloc[0]

    # set parameters tested in experiment
    if infection_type == 'individual' or infection_type == 'threshold':
        params = ['threshold-val', 'percent-grammar-1']
    else:
        params = ['alpha', 'logistic?']

    # display results of experiment for each parameter
    for p in params:
        vals, counts = np.unique(df[p], return_counts=True)
        if len(vals) == 1:
            # if only one value for parameter, no need to plot
            continue

        # check whether effect of tested parameter is statistical relevant to # steps until convergence
        temp_df = df.reset_index(inplace=False).groupby([p, "[run number]"]).max()["[step]"]
        # df stores distribution of # steps until convergence for each parameter value across multiple runs
        exp_result_distr_for_param = temp_df.reset_index(inplace=False).groupby(p).agg({'[step]': ['mean', 'min', 'max', 'std']})
        display(exp_result_distr_for_param)
        plt.plot(exp_result_distr_for_param.index.values, exp_result_distr_for_param['[step]']['mean'], label="mean")
        plt.plot(exp_result_distr_for_param.index.values, exp_result_distr_for_param['[step]']['mean'] + exp_result_distr_for_param['[step]']['std'], label="std", c="r")
        plt.plot(exp_result_distr_for_param.index.values, exp_result_distr_for_param['[step]']['mean'] - exp_result_distr_for_param['[step]']['std'], c="r")
        plt.plot(exp_result_distr_for_param.index.values, exp_result_distr_for_param['[step]']['max'] , c="g", linestyle="--", label="max")
        plt.xlabel(p)
        plt.legend()
        plt.title(infection_type + ': # steps until convergence for different values of ' + p)
        plt.savefig(path2plots + infection_type + '_steps_until_convergence_' + p + '.png', bbox_inches='tight')
        plt.show()

        # display std for different values of p
        plt.figure()
        plt.plot(exp_result_distr_for_param.index.values, exp_result_distr_for_param['[step]']['std'], label="std")
        plt.title(infection_type + ': std of # steps until convergence for different values of ' + p)
        plt.xlabel(p)
        plt.ylabel('std')
        plt.savefig(path2plots + infection_type + '_std_steps_until_convergence_' + p + '.png', bbox_inches='tight')
        plt.show()
        # TODO: docu is wrong, bc std for p=0.9 threshold too high (-> sink-state not wokring for individual)

for experiment_df in dff_all_from_csv:
    # check whether combinatorial experiment worked
    print("new experiment")
    plot_run_length(experiment_df)

```
