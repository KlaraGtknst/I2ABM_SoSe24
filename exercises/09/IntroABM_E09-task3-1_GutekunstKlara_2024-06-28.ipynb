```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
for c_val in [1.0, 2.0]:
    number_individuals = 100
    num_infected = 1
    C = c_val
    infection_history = [num_infected]
    adoption_rate_history = [0]

    while num_infected < number_individuals:
        adoption_rate = np.round(C * num_infected * (number_individuals - num_infected) / number_individuals)
        num_infected += adoption_rate
        infection_history.append(num_infected)
        adoption_rate_history.append(adoption_rate)

    plt.plot(infection_history, label="Infection")
    plt.plot(number_individuals-np.array(infection_history), label="Susceptible")
    plt.title("Infection & Susceptible history with C = " + str(C))
    plt.xlabel("Time")
    plt.legend()
    plt.savefig(f"infection_susceptible_history_c={c_val}.png")
    plt.show()

    plt.plot(adoption_rate_history, label="Rate")
    plt.title("Adoption rate history with C = " + str(C))
    plt.xlabel("Time")
    plt.legend()
    plt.savefig(f"adoption_rate_history_c={c_val}.png")
    plt.show()
```

- bigger C -> faster diffusion (fewer steps to reach 100%)
- bigger C -> adoption rate is strong monoton increasing
